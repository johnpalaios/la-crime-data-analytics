```python
import pyspark
```

```python
from pyspark.sql import SparkSession
```

```python
from pyspark.sql.functions import to_timestamp
from pyspark.sql.functions import *
```

```python
spark = SparkSession \
    .builder \
    .appName("First Query") \
    .config("spark.driver.memory", "2g") \
    .getOrCreate()
```

```python
df = spark.read.parquet("/data/*.parquet")
```

```python
df.createOrReplaceTempView("CrimeData")
```

```python
df.select("Date Rptd").show(5)
```

```python
df_dt = df.select(year(col("Date Rptd")))
```

```python
df_dt.show()
           
```

```python
new_df = df.withColumnRenamed('Date Rptd', 'Date_Rptd')
new_df = new_df.withColumnRenamed('Premis Desc', 'Premis_Desc')
```

```python
new_df.createOrReplaceTempView("CrimeData")
```

```python
spark.sql("SELECT * FROM CrimeData").show(1)
```

```python
spark.sql(""" SELECT Date_Rptd as year FROM CrimeData WHERE Premis_Desc = 'STREET'; """) \
     .show(5)
```
